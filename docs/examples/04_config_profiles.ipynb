# 04 â€“ Using dockyter.toml and profiles

This notebook shows how to:

- create a `dockyter.toml` configuration file,
- use the API backend as the default,
- define reusable Docker profiles with `%docker_profile`.

````python
from pathlib import Path

config_text = """
[backend]
mode = "api"
api_url = "http://127.0.0.1:8000"

[docker]
default_args = "-v /tmp:/tmp ubuntu:22.04"

[profiles]
local = "-v /tmp:/tmp ubuntu:22.04"
ml = "-v /data:/data pytorch/pytorch:latest"
"""

Path("dockyter.toml").write_text(config_text)
print("Wrote config file:", Path("dockyter.toml").resolve())
````

Make sure the example API is running in a terminal:

```bash
uv run uvicorn docs.api_example.server:app --host 127.0.0.1 --port 8000
```

````python
%load_ext dockyter
%docker_status
````

````python
%docker_profile local
!echo "Hello from 'local' profile"
````

````python
%docker_profile ml
!python -c "import torch; print('Torch version:', torch.__version__)"
````
